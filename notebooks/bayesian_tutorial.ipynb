```python
import torch
import torch.nn as nn

from probly.transformation.bayesian import bayesian
```

## Normal Model (No Uncertainty)

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2,1)

    def forward(self, x): 
        return self.linear(x)

model = Net()
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print('Normal output:', model(x))
```

## Apply `bayesian()` -> Get Uncertainty

```python
bayes_model = bayesian(model)
print('New layer:', type(bayes_model.linear).__name__)
```

## Run 3 Times -> See Randomness

```python
torch.manual_seed(42)
for i in range(3):
    mean, std = bayes_model(x)
    print(f'Run {i+1} | Mean: {mean.squeeze().tolist()} | Std: {std.squeeze().tolist()}')
```
